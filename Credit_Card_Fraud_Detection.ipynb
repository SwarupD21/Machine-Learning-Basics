```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
legit=df[df.Class==0]
fraud=df[df.Class==1]
```

**Under Sampling**
Build a sample dataset containing similar distribution of norma transaction and fraudalent transaction

```python
sample_legit = legit.sample(n=492)
```

```python
new_dataset=pd.concat((sample_legit,fraud),axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

**Splitting Data into features and Targets**

```python
X=new_dataset.drop(['Class'],axis=1)
Y=new_dataset['Class']
```

```python
print(Y)
```

**Split the data into trainning and test data**

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2,stratify=Y)
```

**Model Trainning**



```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

**Model Evaluation**

```python
X_train_prediction=model.predict(X_train)
error = metrics.accuracy_score(X_train_prediction,Y_train)
print("Accuracy Score: ",error)
```

```python
X_test_prediction=model.predict(X_test)
error_test = metrics.accuracy_score(X_test_prediction,Y_test)
print("Accuracy Score: ",error_test)
```
