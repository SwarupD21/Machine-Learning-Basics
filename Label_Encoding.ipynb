```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

**Label Encoding for Breast Cancer**

```python
cancer_data=pd.read_csv('/content/data.csv')
```

```python
cancer_data.head()
```

```python
cancer_data['diagnosis'].value_counts()
```

```python
label_encode=LabelEncoder()
```

```python
labels= label_encode.fit_transform(cancer_data.diagnosis)
```

```python
cancer_data['target']=labels
```

```python
cancer_data.head()
```

1-> Maligment

2-> Benian

```python
cancer_data['target'].value_counts()
```

**Label Encoding for Iris Dataset**

```python
iris=pd.read_csv('/content/iris_data.csv')
```

```python
iris.head()
```

```python
iris['Species'].value_counts()
```

```python
labelling=LabelEncoder()
```

```python
label_data=labelling.fit_transform(iris.Species)
```

```python
iris['target']=label_data
```

```python
iris.head()
```

0-> Iris-setosa

1-> Iris-versicolor

2-> Iris-virginica

```python
iris['target'].value_counts()
```
