```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set()
```

```python
plt.quiver(0,0,4,5, scale_units='xy' ,angles='xy', scale=1, color='r')
plt.quiver(0,0,2,1, scale_units='xy' ,angles='xy', scale=1, color='g')
plt.xlim(0,8)
plt.ylim(0,8)
plt.show()
```

**Vector Addition**

```python
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,3,-2])
suming=vector_1+vector_2
print(suming)
```

```python
plt.quiver(0,0,2,3, scale_units='xy' ,angles='xy', scale=1, color='r')
plt.quiver(0,0,3,-2, scale_units='xy' ,angles='xy', scale=1, color='g')
plt.quiver(0,0,5,1, scale_units='xy' ,angles='xy', scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

**Vector Subtraction**

```python
diff=vector_1 - vector_2
print(diff)
```

```python
plt.quiver(0,0,2,3, scale_units='xy' ,angles='xy', scale=1, color='r')
plt.quiver(0,0,3,-2, scale_units='xy' ,angles='xy', scale=1, color='g')
plt.quiver(0,0,-1,5, scale_units='xy' ,angles='xy', scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

**Dot Product**

```python
a=np.array([2,3])
b=np.array([4,4])

a_dot_b= np.dot(a,b)
```

```python
print(a_dot_b)
```

**Cross Product**

```python
a_cross_b=np.cross(a,b)
print(a_cross_b)
```

```python
c=np.array([2,3,4])
d=np.array([5,6,7])
print(np.cross(c,d))
```

**Projection Of Vector**


projection of vector.png

```python
a=np.array([2,5])
v=np.array([8,-6])

magni_of_v = np.sqrt(np.sum(v**2))
proj_of_a_on_v= ((np.dot(a,v))/(magni_of_v**2))*v
print(proj_of_a_on_v)
```
